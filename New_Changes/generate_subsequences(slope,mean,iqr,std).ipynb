```python
# _importing required libraries
import os
import collections

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from scipy.stats import iqr
import statistics

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# _initializing constant variables
input_file_path = os.getcwd() + f'/../data/output_csv/processed_data.csv'
col_names = ['timestamp (s)','subject_id', 'activityID', 'X1', 'Y1', 'Z1', 'X2', 'Y2', 'Z2']
```

```python
# _loading preprocessed data to main dataframe
main_df = pd.read_csv(input_file_path,names=col_names)
main_df
```

```python
#main_df=main_df.head(10000)
```

# Generating subsequences and calculating the statistics for each subsequence

```python
window_length = 30
window_overlap =15
max_window_index = len(main_df)
df=main_df[['subject_id','activityID','X1', 'Y1','Z1','X2','Y2','Z2']]
window_index = 0
Statistics=[]
Activity_ID=[]
Subject_ID=[]
while window_index <= (max_window_index - window_length):
    
    a=df.iloc[window_index:window_index+window_length]
    
    act_ID=a['activityID']
    sub_ID=a['subject_id']
    
    activityID=act_ID[window_index]
    subjectID=sub_ID[window_index]
    
    Activity_ID.append(activityID)
    Subject_ID.append(subjectID)
    

    c=df.values.tolist()
    
    
    #print((window_index+window_length)-1)
    
    if(c[(window_index+window_length)-1][0]-c[window_index][0]==0):
        Acc_slope_y_wrt_x=0
    else:
        Acc_slope_y_wrt_x=(c[(window_index+window_length)-1][1]-c[window_index][1])/(c[(window_index+window_length)-1][0]-c[window_index][0])
        
    if(c[(window_index+window_length)-1][0]-c[window_index][0]==0):
        Acc_slope_z_wrt_x=0
    else:
        Acc_slope_z_wrt_x=(c[(window_index+window_length)-1][2]-c[window_index][2])/(c[(window_index+window_length)-1][0]-c[window_index][0])
        
        
    if(c[(window_index+window_length)-1][2]-c[window_index][2]==0):
        Acc_slope_y_wrt_z=0
    else:
        Acc_slope_y_wrt_z=(c[(window_index+window_length)-1][1]-c[window_index][1])/(c[(window_index+window_length)-1][2]-c[window_index][2])
           
        
    if(c[(window_index+window_length)-1][3]-c[window_index][3]==0):
        gyro_slope_y_wrt_x=0
    else:
        gyro_slope_y_wrt_x=(c[(window_index+window_length)-1][4]-c[window_index][4])/(c[(window_index+window_length)-1][3]-c[window_index][3])
        
    if(c[(window_index+window_length)-1][3]-c[window_index][3]==0):
        gyro_slope_z_wrt_x=0
    else:
        gyro_slope_z_wrt_x=(c[(window_index+window_length)-1][5]-c[window_index][5])/(c[(window_index+window_length)-1][3]-c[window_index][3])
        
        
    if(c[(window_index+window_length)-1][5]-c[window_index][5]==0):
        gyro_slope_y_wrt_z=0
    else:
        gyro_slope_y_wrt_z=(c[(window_index+window_length)-1][4]-c[window_index][4])/(c[(window_index+window_length)-1][5]-c[window_index][5])
        

    statistics=[Acc_slope_y_wrt_x,Acc_slope_z_wrt_x,Acc_slope_y_wrt_z,gyro_slope_y_wrt_x,gyro_slope_z_wrt_x,gyro_slope_y_wrt_z,df['X1'].mean(),df['Y1'].mean(),df['Z1'].mean(),df['X2'].mean(),df['Y2'].mean(),df['Z2'].mean(),df['X1'].std(),df['Y1'].std(),df['Z1'].std(),df['X2'].std(),df['Y2'].std(),df['Z2'].std(),iqr(df['X1']),iqr(df['Y1']),iqr(df['Z1']),iqr(df['X2']),iqr(df['Y2']),iqr(df['Z2'])]   
    
    Statistics.append(statistics)
    
    window_index += window_overlap




```

```python
data=list(zip(Activity_ID,Subject_ID,Statistics))
statistic_df=pd.DataFrame(data,columns=['activity_ID','subject_ID','Statistics'])
statistic_df
```

# Clustering the statistics and generating the centroids of the clusters

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=50,random_state=1234).fit(Statistics)
cluster_ids = pd.DataFrame(model.predict(Statistics), columns=['cluster ID'])
centroids_of_clusters = pd.DataFrame(model.cluster_centers_[cluster_ids['cluster ID']], 
                     columns=['Acc_y_x','Acc_z_x','Acc_y_z','gyro_y_x','gyro_z_x','gyro_y_z','X1_mean','Y1_mean','Z1_mean','X2_mean','Y2_mean','Z2_mean','X1_std','Y1_std','Z1_std','X2_std','Y2_std','Z2_std','X1_iqr','Y1_iqr','Z1_iqr','X2_iqr','Y2_iqr','Z2_iqr'])
centroids_of_clusters
```

```python
centroid_statistic=pd.concat([cluster_ids,centroids_of_clusters],axis=1)
centroid_statistic.drop_duplicates(inplace= True)
centroid_statistic=centroid_statistic.sort_values(by=['cluster ID'])

centroid_statistic['cluster ID'] = centroid_statistic['cluster ID'].astype(str)

```
