```python
def get_label_idx(row):
    
    if sum(row) > 1:
        return complex_activity_id
    
    for idx, val in enumerate(row):
        if val == 1:
            return idx+1
    return 0
```

```python
def generate_words(cluster_cnts):
    
    prefix = 'W_'
    return [prefix+str(val) for val in range(1, cluster_cnts+1)]    
```

```python
def generate_cluster_words(seq_clusters, cluster_words):
    
    # _assign word to each cluster of the subsequence usnig numpy where function
    assigned_words = np.where(seq_clusters != 0, seq_clusters, cluster_words[0])
    for idx in range(1, len(cluster_words)):
         assigned_words = np.where(seq_clusters != idx, assigned_words, cluster_words[idx])
            
    return assigned_words
```

```python
cluster_cnts = 100
doc_df = pd.DataFrame(columns=['timestamp', 'activity_label', 'element_name'])
doc_df['timestamp'] = main_df['timestamp']
doc_df['activity_label'] = main_df['activity_label']
doc_df
```

```python
X = np.array(main_df[sensor_cols])
print(X.shape)
```

```python
cluster_names = generate_words(cluster_cnts)
kmeans = KMeans(n_clusters=cluster_cnts, random_state=0).fit(X)
print('finished clustering..')
doc_df['element_name'] = generate_cluster_words(kmeans.labels_, cluster_names)
doc_df
```

```python
embeddings_filepath = os.getcwd() + f'/../data/extra_sensory_data/word_embeddings_from_clusters.txt'
pd.DataFrame(kmeans.cluster_centers_).to_csv(embeddings_filepath, index=False, header= False)
```

```python
# _save the combined values to text files
for activity in doc_df['activity_label'].unique():
    output_filepath = os.getcwd() + f'/../data/extra_sensory_data/activity_subseq_' + str(activity) + '.txt'
    doc_df.loc[doc_df['activity_label'] == activity][['element_name']].to_csv(output_filepath, sep='\t', index=False, header= False)
```
