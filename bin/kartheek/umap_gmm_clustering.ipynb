```python
import os
import umap
import pandas as pd
import numpy as np

#from sklearn_extra.cluster import KMedoids
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.mixture import GaussianMixture
```

```python
input_file_path = os.getcwd() + f'/../../data/UCIHAR_sensor_features_lstm_tuned_32.csv'
main_df = pd.read_csv(input_file_path, header=None)
main_df.rename(columns={0:'subjectID',1: 'activityID'}, inplace=True)
main_df
```

```python
features = main_df.columns[2:]
scale_df = main_df[features]
scale_df[features] = StandardScaler().fit_transform(scale_df)
vec = scale_df.iloc[:,:].values

model_umap = umap.UMAP(n_neighbors = 15, min_dist = 0.1, n_components = 10, verbose = True)
umap_vec = model_umap.fit_transform(vec)
umap_vec
```

```python
doc_df = pd.DataFrame(columns=['subjectID', 'activityID', 'Features'])
doc_df['subjectID'] = main_df['subjectID'].astype(int)
doc_df['activityID'] = main_df['activityID'].astype(int)
doc_df
```

```python
def get_assigned_words(seq_clusters, cluster_words, doc_df):
    
    # _assign word to each cluster of the subsequence usnig numpy where function
    assigned_words = np.where(seq_clusters != 0, seq_clusters, cluster_words[0])
    for idx in range(1, len(cluster_words)):
         assigned_words = np.where(seq_clusters != idx, assigned_words, cluster_words[idx])
    
    doc_df['Features'] = assigned_words
    
    assigned_clusterWord = pd.DataFrame(data=assigned_words, columns=['cluster_word'])
            
    return assigned_clusterWord
```

```python
def cluster_word_sort(axis_clusters,cluster_names):
    
    result = axis_clusters.loc[(axis_clusters['cluster_word'] == cluster_names)]
    
    return result.iloc[:, 1:]
```

```python
prefix = 'w'
cluster_num = 400
words = [prefix+'_'+str(i) for i in range(cluster_num)]
```

```python
featureVec = StandardScaler().fit_transform(umap_vec)
for val in range(1100, 5000, 100):
  gmm = GaussianMixture(n_components=val, random_state=2).fit(featureVec)
  cluster = gmm.predict(featureVec)
  sc = silhouette_score(featureVec, cluster)
  print(f'cluster count: {val}, sc: {sc}')
```

```python
cluster_centers = []
featureVec = StandardScaler().fit_transform(umap_vec)
model = GaussianMixture(n_components=cluster_num, random_state=2).fit(featureVec)

cluster_ids = pd.DataFrame(model.predict(featureVec), columns=['cluster ID'])
seq_clusters = cluster_ids.to_numpy()
assigned_clusterWord = get_assigned_words(seq_clusters, words)

centroids_of_clusters = pd.DataFrame(model.means_[cluster_ids['cluster ID']])
result = pd.concat([assigned_clusterWord, centroids_of_clusters], axis=1)
result = result.drop_duplicates()
for j in range(len(words)):
    cluster_stats = cluster_word_sort(result,words[j])
    cluster_centers.append(cluster_stats)
cluster_centers
```

```python
embeddings_filepath = os.getcwd() + f'/../../data/sub_sequence_output/word_embeddings_from_clusters.txt'
pd.concat(cluster_centers).to_csv(embeddings_filepath, index=False, header= False)
```

```python
for subject in doc_df['subjectID'].unique():
    
    activity = doc_df.loc[(doc_df['subjectID'] == subject)]['activityID'].values[0]
    output_filepath = os.getcwd() + f'/../../data/documents/train/activity_subseq_' + str(subject) +'_'+ str(activity) + '.txt'
    doc_df.loc[(doc_df['subjectID'] == subject)][['Features']].to_csv(output_filepath, sep='\t', index=False, header= False)
```
