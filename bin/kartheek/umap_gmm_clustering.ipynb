```python
input_file_path = os.getcwd() + f'/../../data/sensor_features_lstm_tuned.csv'
main_df = pd.read_csv(input_file_path, header=None)
main_df.rename(columns={0: 'activityID'}, inplace=True)
main_df
```

```python
features = main_df.columns[1:]
scale_df = main_df[features]
scale_df[features] = StandardScaler().fit_transform(scale_df)
vec = scale_df.iloc[:,:].values

model = umap.UMAP(n_neighbors = 15, min_dist = 0.1, n_components = 2, verbose = True)
umap_vec = model.fit_transform(vec)
umap_vec
```

```python
doc_df = pd.DataFrame(columns=['activityID', 'Features'])
doc_df['activityID'] = main_df['activityID'].astype(int)
doc_df
```

```python
def get_assigned_words(seq_clusters, cluster_words):
    
    # _assign word to each cluster of the subsequence usnig numpy where function
    assigned_words = np.where(seq_clusters != 0, seq_clusters, cluster_words[0])
    for idx in range(1, len(cluster_words)):
         assigned_words = np.where(seq_clusters != idx, assigned_words, cluster_words[idx])
    
    doc_df['Features'] = assigned_words
    
    assigned_clusterWord = pd.DataFrame(data=assigned_words, columns=['cluster_word'])
            
    return assigned_clusterWord
```

```python
def cluster_word_sort(axis_clusters,cluster_names):
    
    result = axis_clusters.loc[(axis_clusters['cluster_word'] == cluster_names)]
    
    return result.iloc[:, 1:]
```
